```python
import matplotlib.pyplot as plt

from wordcloud import (
    WordCloud,
    DescendingFrequencyTable,
    FrequencyData
)
from wordcloud.backends.backend_mpl import CanvasMPL

wc = WordCloud(canvas_backend=CanvasMPL)

c = wc.draw(
    frequency_table=DescendingFrequencyTable([
        FrequencyData('test', 1),
        FrequencyData('this', 10),
        FrequencyData('more', 5),
        FrequencyData('stuff', 5),
        FrequencyData('yeehaw', 5),
        FrequencyData('uwu', 5),
    ]),
    text_props=dict(
        size_max=200,
        rotation_range=(0, 0),
        rotation_step=90,
        rotation_prob=.5
    )
)

plt.imshow(c.data_rgba)
```
