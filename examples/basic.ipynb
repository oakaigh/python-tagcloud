```python
import matplotlib.pyplot as plt

from tagcloud import (
    TagCloud,
    DescendingFrequencyTable,
    FrequencyData,
    graphics
)
from tagcloud.backends.backend_pil import CanvasPIL

wc = TagCloud(canvas_backend=CanvasPIL)

c = wc.draw(
    frequency_table=DescendingFrequencyTable([
        FrequencyData('test', 1),
        FrequencyData('this', 10),
        FrequencyData('more', 5),
        FrequencyData('stuff', 5),
        FrequencyData('yeehaw', 5),
        FrequencyData('uwu', 5),
    ]),
    text_props=dict(
        size_max=200,
        rotation_range=(0, 0),
        rotation_step=90,
        rotation_prob=.5
    ),
    canvas_props=dict(size=graphics.Dimension(600, 800))
    #canvas_props=dict(size=graphics.Dimension(60, 80))
)

_ = plt.imshow(c.data_bilevel)
```
